```python
import cairosvg

def write(size):
    cairosvg.svg2png(
        url='icon.svg',
        write_to='extension/icons/icon'+str(size)+'.png',
        output_width=size,
        output_height=size
    )

write(16)
write(48)
write(128)
```

```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('extension/icons/icon128.png')
plt.imshow(img)
plt.axis('off') # No axes for this plot
plt.show()
```
